```python
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles
from sklearn.datasets import make_moons

from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA
```

```python
x, y = make_circles(n_samples = 500, factor = 0.3, noise = 0.05)

x.shape
```

```python
pd.DataFrame(x).sample(5)
```

```python
y.shape
```

```python
pd.DataFrame(y).sample(5)
```

```python
pl.figure(figsize = (12, 8))
pl.title("Original Space")

pl.plot( x[y == 0, 0], x[y == 0, 1], "yo")
pl.plot( x[y == 1, 0], x[y == 1, 1], "mo")

plt.xlabel("x[0]")
plt.ylabel("x[1]");
```

```python
pca = PCA(n_components = 2)

x_new = pca.fit_transform(x)
```

```python
print("Explained Variance :", pca.explained_variance_ratio_)
```

```python
pl.figure(figsize = (12, 8))
pl.title("PCA Projection")

pl.plot(x_new[y == 0, 0], x_new[y == 0, 1], "yo")
pl.plot(x_new[y == 1, 0], x_new[y == 1, 1], "mo")

plt.xlabel("x_new[0]")
plt.ylabel("x_new[1]");
```

```python
kpca = KernelPCA(kernel = "rbf", gamma = 3)

x_new = kpca.fit_transform(x)

x_new.shape
```

```python
pl.figure(figsize = (12, 8))
pl.title("Kernel PCA")

pl.plot( x_new[y == 0, 0], x_new[y == 0, 1], "yo")
pl.plot( x_new[y == 1, 0], x_new[y == 1, 1], "mo")

plt.xlabel("x_new[0]")
plt.ylabel("x_new[1]");
```
